```python
import pickle
import os


def trigram_converter(trigram): #составляем векторы для мест, в которых потенциально может быть конец предложения
    x = []
    if trigram[0][0].isupper():
        x.append(1)
    else:
        x.append(0) 
    if trigram[2][0].isupper():
        x.append(1)
    else:
        x.append(0)  
    if trigram[0].isupper():
        x.append(1)
    else:
        x.append(0)  
    if trigram[2].isupper():
        x.append(1)
    else:
        x.append(0)  
    x.append(len(trigram[0])) 
    x.append(len(trigram[2]))  
    if trigram[2] in ['$', '£', '€']:  
        x.append(1)
    else:
        x.append(0)
    if trigram[2].isdigit(): 
        x.append(1)
    else:
        x.append(0)
    print(x) 
    return x

def sentence_tokenizer(text, classifierpath):
    classifier = load_classifier(classifierpath)
    sentences = []
    sentence=''
    #text = text.replace('...', '')
    text = text.replace('..', '')
    text = text.replace('. .', '')
    #print(text)
    dot_splitted_text = text.split('.')#массив из разделенных потенциальных предложений
    for idx, seg in enumerate(dot_splitted_text):
        if idx < len(dot_splitted_text) - 2:
            part1 = str.strip(seg).split(' ')[-1]
            part2 = str.strip(dot_splitted_text[idx + 1]).split(' ')[0]
            boundary_trigram = part1 + '\t.\t' + part2
            convert_vector = trigram_converter(boundary_trigram.split())
            prediction = classifier.predict([convert_vector])
            #print(prediction)
            if prediction == [1] :
                print('Найдено новое предложение')
                sentence += seg+'.'
                sentences.append(sentence)
                sentence = ''
            else:
                sentence += seg+'.'
        else:
            sentence += seg
            sentences.append(sentence)
            sentence = ''
    for s in sentences:
        print(s)
    return sentences


def load_classifier(path):
    modelfile = open(path, 'rb')
    return pickle.load(modelfile, encoding='latin1')


if __name__ == '__main__':
    text_rus2  = 'Общество с ограниченной ответственностью "Мебель-радиотовары" (далее - ООО "Мебель-радиотовары") обратилось в Арбитражный суд Оренбургской области с иском к муниципальному унитарному предприятию "Старая Русь" (далее - МУП "Старая Русь"), обществу с ограниченной ответственностью "Меркурий+" (далее - ООО "Меркурий+"), обществу с ограниченной ответственностью "Авангард" (далее - ООО "Авангард"), Управлению Федеральной регистрационной службы по Оренбургской области (далее - управление), администрации г. Орска о признании недействительными в силу ничтожности притворных договоров купли-продажи от 29.04.2003 N 7 и от 25.08.2003 и применении последствий недействительности ничтожных сделок в виде возврата ООО "Авангард" ООО "Меркурий+" нежилого встроенного помещения N 1 полезной площадью 240,9 кв. м (площадь подвала 158,7 кв. м), расположенного на первом этаже и в подвале 3- 4-этажного шлакоблочного жилого дома по адресу: Оренбургская обл., г. Орск, пр. Ленина, д. 53; признания недействительной регистрационной записи в Едином государственном реестре прав на недвижимое имущество и сделок с ним о государственной регистрации перехода права собственности за ООО "Авангард" на спорное нежилое помещение; возврата ООО "Авангард" ООО "Меркурий+" 1110000 руб.; возврата ООО "Меркурий+" МУП "Старая Русь" спорного нежилого помещения; возврата МУП "Старая Русь" ООО "Меркурий+" 1100000 руб.; третье лицо - Комитет по управлению имуществом г. Орска.Решением суда первой инстанции от 20.02.2006 (судья Горохов В.А.) в удовлетворении иска в отношении администрации г. Орска и Управления Федеральной регистрационной службы отказано. В остальной части исковые требования удовлетворены полностью: договоры купли-продажи от 29.04.2003 N 7 и от 25.08.2003 признаны недействительными. ООО "Авангард" обязано возвратить ООО "Меркурий+" помещение, регистрационная запись о государственной регистрации за ООО "Авангард" на помещение признана недействительной. ООО "Авангард" обязано возвратить ООО "Меркурий+" 1100000 руб. ООО "Меркурий+" обязано возвратить МУП "Старая Русь" помещение, МУП "Старая Русь" обязано возвратить ООО "Меркурий+" 1100000 руб.В порядке апелляционного судопроизводства решение не пересматривалось.В жалобе, поданной в Федеральный арбитражный суд Уральского округа, заявитель - ООО "Авангард" - просит решение суда первой инстанции отменить и принять новый судебный акт об отказе в удовлетворении, ссылаясь на нарушение судом норм, предусмотренных ст. 16, 69 Арбитражного процессуального кодекса Российской Федерации, ст. 167 Гражданского кодекса Российской Федерации, п. 2 ст. 3 Федерального закона "О приватизации государственного и муниципального имущества".'
    text_rus1 = 'Данный вывод суда следует признать ошибочным.Постановлением Федерального арбитражного суда Уральского округа от 15.09.2004 по делу N Ф09-3039/04-ГК (N А47-8286/03) установлено, что распоряжение от 15.12.2002 N 3746-р принято в пределах полномочий администрации г. Орска, не противоречит действующему законодательству и не нарушает прав и охраняемых законом интересов ООО "Мебель-радиотовары".'
    text_rus3 ='Лица, участвующие в деле, о времени и месте рассмотрения кассационной жалобы извещены надлежащим образом путем направления в их адреса копий определения о принятии кассационной жалобы к производству заказным письмом с уведомлением, а также размещения данной информации на официальном сайте Федерального арбитражного суда Уральского округа, явку своих представителей в судебное заседание не обеспечили.От индивидуального предпринимателя Гизетдинова Шамиля Закариевича в суд кассационной инстанции поступило ходатайство о рассмотрении дела в его отсутствие. Ходатайство судом кассационной инстанции рассмотрено и удовлетворено в соответствии с ч. 3 ст. 284 Арбитражного процессуального кодекса Российской Федерации.'
    
    #text_rus1, text_rus2, text_rus3 - Тестовые примеры, которые можно подставить в sentence_tokenizer в следующей строке, вообще в идеальном мире туда можно подставить любой текст и разбиение будет верным
    sents = sentence_tokenizer(text_rus1, os.path.abspath('../data/SVM_Classifier'))
```
